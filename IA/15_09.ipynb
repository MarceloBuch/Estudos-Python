```python
def calcMedia(lista):
  soma = 0
  for nota in lista:
    soma = soma + nota
  return soma/len(lista)
Lista_notas= [3.4,6.6,8,9,10,9.5,8.8,4.3]
print("Media =", calcMedia(Lista_notas))
```

```python
def categoria(idade):
  if idade < 3:
    ctg = 'Bebe'
  elif idade < 10:
    ctg = 'Infantil'
  elif idade < 14:
    ctg = 'Jubior'
  elif idade < 18:
    ctg = 'Adolescente'
  elif idade < 30:
     ctg = 'Jovem'
  else:
     ctg = 'Adulto'
  return ctg

idade = int(input('Digite a idade: '))
print("A categoria é",categoria(idade))
```

```python
def incluir(d):
  nome = input("Digite o nome a incluir:")
  if nome in d:
    print("Nome já existente")
  else:
    fone = input("Digite o telefone:")
    d[nome] = fone

def consultar(d):
  nome = input("Digite o nome a consultar: ")
  print("Telefone: ", d.get(nome, "Não Disponível"))

def alterar(d):
  nome = input("Digite o nome do fone a alterar: ")
  print("Telefone atual: ", d.get(nome, "Não diponivel"))
  if nome in d:
    fone = input("Digite o novo telefone: ")
    d[nome] = fone
    print("Telefone alterado")
  else:
    print("Nome não encontrado para alteração")

def excluir(d):
  nome = input("Digite o nome a ser excluido: ")
  if nome in d:
    del d[nome]
    print("Nome e telefone excluídos")
  else:
    print("Nome não encontrado")

def visualizar(d):
  print("Agenda: ")
  for nome in d.keys():
    print(nome, d[nome])

agenda = {}

incluir(agenda)
incluir(agenda)
incluir(agenda)
consultar(agenda)
alterar(agenda)
visualizar(agenda)
```

```python
agenda = {}
opc = 10
while opc != 0:
  opc = int(input("0 - sair, 1 - incluir, 2 - consultar, 3 - alterar, 4 - excluir, 5 - visualizar"))
  if opc == 1:
    incluir(agenda)
  elif opc == 2:
    consultar(agenda)
  elif opc == 3:
    alterar(agenda)
  elif opc == 4:
    excluir(agenda)
  elif opc == 5:
    visualizar(agenda)
  else:
    print("opção inválida")
```
