```python
from sklearn.datasets import load_iris
from sklearn import tree
from sklearn.model_selection import train_test_split
import pandas as pd
import sklearn.metrics as metrics
```

```python
iris = load_iris()
print(iris['DESCR'])
```

```python
X_train, X_test, y_train, y_test = train_test_split(iris.data, iris.target, test_size=0.25)
```

```python
clf = tree.DecisionTreeClassifier(criterion='entropy') #quando entropia é igual a zero significa que a arvore encontrou uma predição para os testes
clf = clf.fit(X_train, y_train)
predictions = clf.predict(X_test)
```

```python
print('\n Matriz de confusão\n', pd.crosstab(y_test, predictions, rownames=['Real'], colnames=['Predito'], margins = True, margins_name="Todos"))
```

```python
print('Relatório sobre a qualidade: \n')
print(metrics.classification_report(y_test, predictions, target_names = ['Setosa', 'Versicolor', 'Virginica']))
```
